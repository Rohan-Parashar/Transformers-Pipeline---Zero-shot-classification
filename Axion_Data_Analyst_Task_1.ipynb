```python
import pandas as pd
from transformers import pipeline

# Initialize the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Define your candidate labels
candidate_labels = taxo_task_1[' Sentiment'][:11].tolist()

# Create a list to store the results
results = []

# Loop through all entries in raw_data_t1['Review']
for review_text in raw_data_t1['Review'][:31]:
    # Get the top 2 sentiment labels
    result = classifier(review_text, candidate_labels, top_k=2)

    # Get the top 2 labels and their scores
    top_labels = result['labels'][:2]

    # Append the review and top labels to the results list
    results.append({
        'Review': review_text,
        'Sentiment 1': top_labels[0],  # Label with the highest score
        'Sentiment 2': top_labels[1]   # Label with the second highest score
    })

# Convert the results list into a DataFrame
df_sentiments = pd.DataFrame(results)

# Display the DataFrame
print(df_sentiments)
```

```python
# # Symptom Conditions

# Initialize the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Define your candidate labels
candidate_labels = taxo_task_1['Symptom Condition'][:38].tolist()

# Create a list to store the results
results = []

# Loop through all entries in raw_data_t1['Review']
for review_text in raw_data_t1['Review'][:31]:
    # Get the top 3 symptom conditions
    result = classifier(review_text, candidate_labels, top_k=3)

    # Get the top 3 labels and their scores
    top_labels = result['labels'][:3]

    # Append the review and top labels to the results list
    results.append({
        'Symptom Condition 1': top_labels[0],  # Label with the highest score
        'Symptom Condition 2': top_labels[1],  # Label with the second highest score
        'Symptom Condition 3': top_labels[2]   # Label with the third highest score
    })

# Convert the results list into a DataFrame
df_symptom_condition = pd.DataFrame(results)

# Display the DataFrame
print(df_symptom_condition)
```

```python
# #  Symptom Components

# Initialize the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Define your candidate labels
candidate_labels = taxo_task_1['Symptom Component'][:21].tolist()

# Create a list to store the results
results = []

# Loop through all entries in raw_data_t1['Review']
for review_text in raw_data_t1['Review'][:31]:
    # Get the top 3 symptom component
    result = classifier(review_text, candidate_labels, top_k=3)

    # Get the top 3 labels and their scores
    top_labels = result['labels'][:3]

    # Append the review and top labels to the results list
    results.append({
        'Symptom Component 1': top_labels[0],  # Label with the highest score
        'Symptom Component 2': top_labels[1],
        'Symptom Component 3': top_labels[2]# Label with the second highest score
    })

# Convert the results list into a DataFrame
df_symptom_component = pd.DataFrame(results)

# Display the DataFrame
print(df_symptom_component)
```

```python
# #  Failure Condition

# Initialize the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Define your candidate labels
candidate_labels = taxo_task_1['Failure Condition'][:12].tolist()

# Create a list to store the results
results = []

# Loop through all entries in raw_data_t1['Review']
for review_text in raw_data_t1['Review'][:31]:
    # Get the top 3 Failure Condition
    result = classifier(review_text, candidate_labels, top_k=3)

    # Get the top 3 labels and their scores
    top_labels = result['labels'][:3]

    # Append the review and top labels to the results list
    results.append({
        'Failure Condition 1': top_labels[0],  # Label with the highest score
        'Failure Condition 2': top_labels[1],  # Label with the second highest score
        'Failure Condition 3': top_labels[2]   # Label with the third highest score
    })

# Convert the results list into a DataFrame
df_failure_condition = pd.DataFrame(results)

# Display the DataFrame
print(df_failure_condition)
```

```python
# #  Failure Components

# Initialize the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Define your candidate labels
candidate_labels = taxo_task_1['Failure Component'][:18].tolist()

# Create a list to store the results
results = []

# Loop through all entries in raw_data_t1['Review']
for review_text in raw_data_t1['Review'][:31]:
    # Get the top 3 Failure Component
    result = classifier(review_text, candidate_labels, top_k=3)

    # Get the top 3 labels and their scores
    top_labels = result['labels'][:3]

    # Append the review and top labels to the results list
    results.append({
        'Failure Component 1': top_labels[0],  # Label with the highest score
        'Failure Component 2': top_labels[1],  # Label with the second highest score
        'Failure Component 3': top_labels[2]   # Label with the third highest score
    })

# Convert the results list into a DataFrame
df_failure_component = pd.DataFrame(results)

# Display the DataFrame
print(df_failure_component)
```
